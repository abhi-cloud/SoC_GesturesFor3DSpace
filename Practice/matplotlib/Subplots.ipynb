<li>
    Two Axes embeded in same figure
</li>

```python
from matplotlib import pyplot as plt
import pandas as pd

plt.style.use('seaborn')

data = pd.read_csv('data2.csv')
ages = data['Age']
all_devs = data['All_Devs']
python = data['Python']
js = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, all_devs, color='#666666', linestyle='--', label='All_Devs')

ax2.plot(ages, python, label='Python')
ax2.plot(ages, js, label='JavaScript')

ax1.legend()
ax1.set_title('Medaian Salaries')
ax1.set_ylabel('Salaries (in USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Salaries (in USD)')

plt.tight_layout()
plt.show()
```

<li>
    Two different figures
</li>

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, all_devs, color='#666666', linestyle='--', label='All_Devs')

ax2.plot(ages, python, label='Python')
ax2.plot(ages, js, label='JavaScript')

ax1.legend()
ax1.set_title('Medaian Salaries')
ax1.set_ylabel('Salaries (in USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Salaries (in USD)')

plt.tight_layout()
plt.show()
```
