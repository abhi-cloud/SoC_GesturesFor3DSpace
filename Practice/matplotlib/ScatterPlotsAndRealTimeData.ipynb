## Using scatter plots we can analyse muti-dimensional data using 2D images
We can set size and colour properties as well as the coordinates

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('data3.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.7)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike ratio')

plt.xscale('log')      # marking in lograthmic scale on both axes
plt.yscale('log')

plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.title('YouTube Videos\' trends')
# plt.legend()

plt.show()
```

<h1 style='font-size:30px; font-family:FreeMono'>TimePlot</h1>

```python
from datetime import datetime, timedelta
from matplotlib import dates as m_dates

plt.style.use('fivethirtyeight')

dates = [
    datetime(2020, 3, 15),
    datetime(2020, 3, 29),
    datetime(2020, 3, 31),
    datetime(2020, 4, 10),
    datetime(2020, 4, 12),
    datetime(2020, 5, 5),
    datetime(2020, 5, 11),
]

y = [3, 7, 5, 4, 9, 2, 1]

plt.plot_date(dates, y, linestyle='solid', linewidth=2)
plt.gcf().autofmt_xdate()

date_format = m_dates.DateFormatter('%b %d')
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

```python
from itertools import count
import random
from matplotlib.animation import FuncAnimation

x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    plt.cla()
    plt.plot(x_vals, y_vals)
    
ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()
```
